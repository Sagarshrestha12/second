```python
#creating the numpy array by several methods
import numpy as np
# we can create the numpy array using array() in numpy object
a=np.array([[1,2,3],(1,22,3)],dtype=np.int32)# it should have same parameter of the same data type of type list
print(a)
# array funtion always take list as an argument 
# Numpy array occupies less memory as compare to list

# To create sequence of the numbers,Numpy provides the funtion analogous to range that return array insted of list 
# It is arrange()
c=np.arange(0,20,0.3)# it also accepts float increment
print(c)
d=list(range(20))
print(c)

# type of the array can be specified explicitly by passing the parameters of dtype 
c=np.array([[1,2],[3,4]],dtype=complex)
print(c)

# There is many specail funtion in numpy module, it consists of the ones() ,zeros() which is used create an numpy array 
# of zero and one respectively
# for example to create an numpy array containing zero can be made for zeros() contained in numpy module
x=np.zeros((3,4))# where 3 are the no. of rows and 4 are no. of coloums
print(x)
# np.dtype returns the type of the elements in array
print(x.ndim)
print(x.itemsize) # return the size in bytes of each elements in array

# linspace() recieves the number of the elements that we want instead of the steps followed
# it useful than arange() 
x=np.linspace(0,2,9)# where 9 is no of elements obtained between the 0 and 2
print(x)

# concatination of the numpy array
x=np.array([[2,3],[3,4]])
y=np.array([[4,5],[7,8]])
z=np.concatenate((x,y),axis=0);# concatination example of numpy array x and y along rows since axis=0(i.e no of rows incr.)
z1=np.concatenate((x,y),axis=1);# concatenation example of numpy array x and y along coloums since axis=1 
print("\n contenation operation on x and y numpy array\n",z,"\ndsfa",z1)
```
